```python
import hashlib
password = input("Enter password: ")
hashed = hashlib.sha256(password.encode()).hexdigest()
print("Hashed password:", hashed)
```

```python
import hashlib
while True:
    password = input("Enter password or type exit: ")
    if password == "exit":
        break
    # Strength check
    if len(password) < 8:
        print("Weak password")
    else:
        print("Stronger password")
    # Hashing
hashed = hashlib.sha256(password.encode()).hexdigest()
print(hashed)
```
